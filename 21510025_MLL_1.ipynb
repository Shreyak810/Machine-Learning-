11.Display houses having number of bedrooms more than 2 and square foot

```python
filtered_houses = df[(df['bedrooms'] > 2) & (df['sqft_living'] > 2000)]
count_filtered_houses = filtered_houses.shape[0]
print("Houses with > 2 bedrooms and square footage > 2000:")
print(filtered_houses)
print("Count:", count_filtered_houses)
```

12.Replace/eliminate missing values

```python
print("Remove rows with any missing values")
print(df.dropna(axis=0, how='any', inplace=True))

print("Remove column with any missing values")
print(df.dropna(axis=1, how='any', inplace=True))
```

13.Change column name(s) to short/easy names if required.

```python
df.rename(columns={'sqft_lot': 'sqftlot', 'sqft_above': 'sqftabove'}, inplace=True)
print("Change in name of columns")
print("Column name: ")
print(df.columns);
```

14.Drop unessential columns (feature selection).

```python
unessential_columns = ['waterfront', 'long']
df.drop(columns=unessential_columns, inplace=True)
print("Column name: ")
print(df.columns)
```

15.Find mean/min/max of numeric columns.

```python
df_numeric = df.apply(pd.to_numeric, errors='coerce')
mean_values = df_numeric.mean()
min_values = df_numeric.min()
max_values = df_numeric.max()

print("Mean values:")
print(mean_values)

print("\nMinimum values:")
print(min_values)

print("\nMaximum values:")
print(max_values)
```

17.Display unique values in each column

```python
for column in df.columns:
    unique_values = df[column].unique()
    print(f"Unique values in '{column}':")
    print(unique_values)
    print()
```
