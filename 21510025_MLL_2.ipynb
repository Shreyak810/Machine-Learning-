**Q4**.	Plot histogram of every numerical attribute and mention your observations in terms of normal behavior, skewness and kurtosis.

**NOTE**
1. Normal Behavior:bell-shaped curve-->normal behavior
2. Skewness: measures the asymmetry of the distribution.
Skewness > 0: Right-skewed (tail extends more to the right).
Skewness < 0: Left-skewed (tail extends more to the left).
Skewness = 0: Symmetric (no skew).
3. Kurtosis:Kurtosis measures the tails and sharpness of the peak of the distribution.
Kurtosis > 0: Leptokurtic (heavy-tailed distribution with a sharp peak).
Kurtosis < 0: Platykurtic (light-tailed distribution with a flat peak).
Kurtosis = 0: Mesokurtic (normal distribution).


**1.Normal Behavior:**
Symmetry: If the histogram is symmetric, it suggests a normal distribution. This means that data is evenly distributed on both sides of the mean.
Skewness: If the histogram is skewed to the right (positively skewed), the tail is longer on the right side. If it's skewed to the left (negatively skewed), the tail is longer on the left side.

**2.Skewness:**
Positive Skewness: If the right tail is longer, the distribution is positively skewed. The mean is greater than the median.
Negative Skewness: If the left tail is longer, the distribution is negatively skewed. The mean is less than the median.

**3.Kurtosis:**
Leptokurtic: If the peak of the histogram is higher and sharper than a normal distribution, it is leptokurtic (positive kurtosis). Tails are fatter, and there are more extreme values.
Mesokurtic: If the peak of the histogram is similar to a normal distribution, it is mesokurtic (normal kurtosis).
Platykurtic: If the peak is flatter than a normal distribution, it is platykurtic (negative kurtosis). Tails are thinner, and there are fewer extreme values.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot histograms for all numerical attributes
for column in df1.select_dtypes(include='number').columns:
    plt.figure(figsize=(6,5)) # sets the size of the figure (width and height) for each histogram using matplotlib.pyplot.figure.
    sns.histplot(df1[column], kde=True) # uses Seaborn's histplot function to create a histogram for the selected column. The kde=True parameter adds a Kernel Density Estimate (a smooth curve) to the plot.
    plt.title(f'Histogram of {column}')
    plt.show()
```

```python
from scipy.stats import skew, kurtosis

for column in df1.select_dtypes(include='number').columns:
    print(f'{column}:')
    print(f'Skewness: {skew(df1[column])}')
    print(f'Kurtosis: {kurtosis(df1[column])}')
```

7.	Perform standardization and normalization of the numerical features and analyze the transformation visually using suitable visualization.


**NOTE**

*Why Normalize:*

1.   Ensures that all features have the same
2.   scale, preventing certain features from dominating others.
Useful when features have different ranges and units.

---



*Why Standardize:*


1.   Centers the data around zero, making it easier to interpret.
2.   Useful when features have different means and standard deviations.


The fit_transform method computes the mean and standard deviation from the data and then standardizes the columns accordingly.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler

df = pd.read_csv('AirQualityUCI2.csv', sep=';')
numerical_columns = df.select_dtypes(include=['number']).columns
missing_values = df[numerical_columns].isnull().sum()
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].mean())


try:
    standard_scaler = StandardScaler()#Create an instance of the StandardScaler class
    df_standardized = df.copy()#To avoid modifying original data
    df_standardized[numerical_columns] = standard_scaler.fit_transform(df[numerical_columns])
    #The fit_transform method computes the mean and standard deviation from the data and then standardizes the columns accordingly.

    min_max_scaler = MinMaxScaler()
    df_normalized = df.copy()
    df_normalized[numerical_columns] = min_max_scaler.fit_transform(df[numerical_columns])


    for column in numerical_columns:
        plt.figure(figsize=(15, 5))

        # Original Data
        plt.subplot(1, 3, 1)
        sns.histplot(df[column], kde=True)
        plt.title(f'Original - {column}')

        # Standardized Data
        plt.subplot(1, 3, 2)
        sns.histplot(df_standardized[column], kde=True)
        plt.title(f'Standardized - {column}')

        # Normalized Data
        plt.subplot(1, 3, 3)
        sns.histplot(df_normalized[column], kde=True)
        plt.title(f'Normalized - {column}')

        plt.tight_layout()
        plt.show()

except Exception as e:
    print(f"Error during standardization/normalization: {e}")
```

8.	Perform dimensionality reduction using PCA and t-SNE in 2 dimensions. Compare visual plots and note your observations.

**NOTES**


1.   Principal Component Analysis (PCA)



*   Linear Structure
*   Computational Efficiency: PCA is computationally efficient, especially for large datasets.
*   Noise Reduction: PCA tends to be less sensitive to noisy data compared to t-SNE. It can be helpful when there is a significant amount of noise in the data.






2.   t-distributed Stochastic Neighbor Embedding (t-SNE)

*   Non-Linear Structure

*  Small to Medium-Sized Datasets:
*   Preservation of Clusters


*   Local Relationships








```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_excel('AirQualityUCI.xlsx')

X = df[['CO(GT)', 'PT08.S1(CO)', 'NMHC(GT)', 'C6H6(GT)', 'PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)', 'PT08.S5(O3)']]

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Apply t-SNE
tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X_scaled)

# Visualize the results
plt.figure(figsize=(12, 6))

# Plotting PCA results
plt.subplot(1, 2, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.title('PCA Visualization')

# Plotting t-SNE results
plt.subplot(1, 2, 2)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1])
plt.title('t-SNE Visualization')

plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_excel('AirQualityUCI.xlsx')

X = df[['T','RH','AH','CO(GT)', 'PT08.S1(CO)', 'NMHC(GT)', 'C6H6(GT)', 'PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)', 'PT08.S5(O3)']]

# Standardize the data----->Standardizes the selected features using StandardScaler to ensure mean=0 and variance=1.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Apply t-SNE
tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X_scaled)

# Visualize the results
plt.figure(figsize=(12, 6))

# Plotting PCA results
#x-axis represents the first principal component (X_pca[:, 0])
#and the y-axis represents the second principal component (X_pca[:, 1])
plt.subplot(1, 2, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.title('PCA Visualization')

# Plotting t-SNE results
plt.subplot(1, 2, 2)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1])
plt.title('t-SNE Visualization')

plt.show()
```

9.	If m denotes number of features and n denotes number of data points, for n = 10000, create following datasets using randn function in python.
a.	m = 10
b.	m = 50
c.	m = 100
d.	m = 500
e.	m = 1000
f.	m = 1500
Plot m vs execution time for PCA and t-SNE and note your observations. You are free to experiment with many more m values. (use timeit module in python for computing execution time.)



**NOTES**
1.   **timeit.timeit**:used to measure the execution time of a given statement or function.

*   stmt (statement): The code whose execution time needs to be measured.

*   number: The number of times the code should be executed. In this case, number=1 means the code is executed only once.







```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import timeit
import matplotlib.pyplot as plt

# Creating a random dataset
def generate_dataset(n, m):
    return np.random.randn(n, m)

# Computing PCA execution time
def pca_execution_time(data):
    pca = PCA(n_components=2)
    return timeit.timeit(lambda: pca.fit_transform(data), number=1)

# Computing t-SNE execution time
def tsne_execution_time(data):
    tsne = TSNE(n_components=2)
    return timeit.timeit(lambda: tsne.fit_transform(data), number=1)


m_values = [10, 50, 100, 500, 1000, 1500]
n = 10000
pca_times = []
tsne_times = []

for m in m_values:
    data = generate_dataset(n, m)

    #For PCA
    pca_time = pca_execution_time(data)
    pca_times.append(pca_time)

    # For t-SNE
    tsne_time = tsne_execution_time(data)
    tsne_times.append(tsne_time)

plt.figure(figsize=(10, 6))
plt.plot(m_values, pca_times, label='PCA')
plt.plot(m_values, tsne_times, label='t-SNE')
plt.xlabel('Number of Features (m)')
plt.ylabel('Execution Time (seconds)')
plt.title('PCA vs t-SNE Execution Time')
plt.legend()
plt.show()
```
